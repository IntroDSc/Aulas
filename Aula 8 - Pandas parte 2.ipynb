As boas práticas em Python recomendam que a carga dos pacotes necessários seja feita no início do script. 

```python
import pandas as pd
import numpy as np
```

Vamos carregar a base de dados Titanic da internet.

```python
# Dados da base Titanic obtidos em https://gist.github.com/fyyying:
url = "https://gist.githubusercontent.com/fyyying/4aa5b471860321d7b47fd881898162b7/raw/6907bb3a38bfbb6fccf3a8b1edfb90e39714d14f/titanic_dataset.csv"

dados = pd.read_csv(url)
dados.head(3)
```

# Geração e deleção de colunas

`pandas` permite que criemos colunas a partir de outras colunas, ou diretamente, fonecendo os valores.

```python
df_1classe = dados.loc[dados["Pclass"] == 1, 
                       ["Survived", "Sex", "Age", "SibSp", "Parch"]]
df_1classe
```

```python
df_1classe["Acompanhantes"] = df_1classe["SibSp"] +  df_1classe["Parch"]
df_1classe
```

```python
ids = ["Passageiro_" + str(i) for i in range(216)]

df_1classe["ID"] = ids
df_1classe
```

A deleção de colunas é feita com o método `drop`, definindo o argumento *axis* = 1. Caso ele seja definido para 0, a deleção é aplicada a linhas. Em vez de usar os argumentos *labels* e *axis*, pode-se usar os argumentos *columns* e *index*.

```python
df_1classe.drop(labels = ["ID"], axis = 1)
```

```python
df_1classe.drop(columns = ["ID"])
```

Para renomear colunas, usamos o método `rename`.

```python
df_1classe.rename(
    columns = {
        "Survived": "Sobrevivente",
        "Sex": "Sexo",
        "Age": "Idade",
        "SibSp": "Irmãos_Esposos",
        "Parch": "Pais_Filhos",
    },
    inplace = True)
df_1classe
```

Em vez de passar um dicionário para o argumento `columns`, podemos passar uma função que gere como resultado um conjunto de cadeias de caracteres.

```python
df_1classe.rename(columns = str.lower, inplace = True)
df_1classe
```

# Ordenação por valores

A ordenação de linhas de *DataFrames* em função dos valores contidos em suas colunas é feita por meio do método `sort_values`.

```python
df_1classe.head()
```

# Agrupamento de linhas

Dados podem ser agrupados por seus valores para a geração de estatísticas descritivas e a aplicação de funções. O processo se inicia com o método `groupby`. Em seguida, geram-se os resultados desejados com outros métodos ou funções.

```python
dados.head()
```

```python
dados. \
    groupby(by = ["Survived", "Sex", "Pclass"],
            dropna = True). \
    mean(). \
    round(2)
```

Note que o resultado incluiu todas as colunas que pudessem ser tratadas pela função ou método utilizado (`mean`). Podemos escolher os resultados a serem apresentados após o cálculo ser feito usando `[]`, `loc` ou `iloc`.

```python
dados. \
    groupby(by = ["Survived", "Sex", "Pclass"],
            dropna = True)[["Age", "Fare"]]. \
    mean(). \
    round(2)
```

```python
dados. \
    groupby(by = ["Survived", "Sex", "Pclass"],
            dropna = True). \
    mean(). \
    round(2). \
    loc[:, ["Age", "Fare"]]
```

Na expressão acima, vale também destacar o uso do argumento *dropna*, que permite que linhas com valores omitidos (`NaN`) numa coluna usada para o agrupamento sejam desconsideradas.

O método `count` é útil para totalizar o número de registros que caem em cada subgrupo.

```python
dados. \
    loc[:, ["Survived", "Sex", "Pclass", "Name"]]. \
    groupby(by = ["Survived", "Sex", "Pclass"], 
            dropna = True). \
    count()
```

O mesmo resultado pode ser obtido com o método `value_counts`. No entanto, é necessário usar o método `to_frame` para converter o objeto gerado para um *DataFrame*.

```python
dados. \
    loc[:, ["Survived", "Sex", "Pclass"]]. \
    value_counts(). \
    to_frame(). \
    rename(columns = {0: "Total"})
```

O método `agg` também pode ser usado em objetos do tipo *GroupBy*. A seguir, usamos a funcionalidade [named aggregations](https://pandas.pydata.org/pandas-docs/stable/user_guide/groupby.html#groupby-aggregate-named) disponível a partir da versão 0.25 do `pandas` para definir o nome das colunas geradas.

```python
import numpy as np

dados \
    .loc[:, ["Survived", "Sex", "Pclass", "Age"]] \
    .groupby(by = ["Sex", "Pclass"], 
             dropna = True) \
    .agg(
         Sobreviventes = ("Survived", 'sum'),
         Percentual = ("Survived", lambda x: str(round(100 * np.mean(x))) + "%"),
         Média_de_Idade = ("Age", lambda x: round(np.mean(x), 1))
        )
```

Como visto acima, `sum` e `mean` podem ser usados para calcular a quantidade e a proporção, respectivamente, de valores 1 em variáveis categóricas codificadas com valores 0 e 1 (coluna "Survived").

É importante destacar a estrutura hierarquizada de índices nos resultados gerados com `groupby`. O objeto gerado para a indexação do *DataFrame* é do tipo ***MultiIndex***.

```python
props = dados. \
    loc[:, ["Survived", "Sex", "Pclass"]]. \
    groupby(by = ["Sex", "Pclass"], 
            dropna = True). \
    mean(). \
    round(2)
props
```

```python
props.index
```

Abaixo, o uso do método `describe` em um objeto do tipo *GroupBy* gera índices do tipo *MultiIndex* tanto para as linhas, quanto para as colunas.

```python
descritivas = dados. \
    groupby(by = ["Survived", "Sex", "Pclass"], 
            dropna = True). \
    describe(). \
    round(1)
descritivas
```

```python
descritivas.columns
```

A função `IndexSlice` permite que façamos seleções de partes de índices hierárquizados de forma mais intuitiva.

```python
descritivas.loc[pd.IndexSlice[:, ["female"], [1, 3]], 
                pd.IndexSlice[["Age", "Fare"], ["mean", "std"]]]
```

A página [MultiIndex / advanced indexing](https://pandas.pydata.org/pandas-docs/stable/user_guide/advanced.html) provê informações detalhadas sobre a criação e tratamento de hierarquias de índices.

# Concatenação e junções

*DataFrames* podem ser concatenados linha a linha, ou coluna a coluna.

```python
df1 = pd.DataFrame(
    {
        "Ano": [2020, 2020, 2020, 2021, 2021],
        "Semestre": [1, 1, 2, 1, 2],
        "Nome": ["José Martins", "Marcia Carneiro", "Joana Tenório", "Carlos Mendes",
                 "Victor Lunes"]
    })
df2 = pd.DataFrame(
    {
        "Ano": [2021, 2021],
        "Semestre": [2, 2],
        "Nome": ["Ana Maria Braga", "Ricardo Conceição"]
    })
```

```python
df1
```

```python
df2
```

Quando as colunas têm os mesmos nomes, a concatenação de linhas pode ser feita usando o método `append`, ou o método `concat`.

```python
df1.append(df2)  # Gera uma cópia com as linhas adicionais
```

```python
df1
```

```python
pd.concat([df1, df2])  # Também gera uma cópia
```

Nas operações acima, os índices originais foram mantidos. Assim, o *DataFrame* gerado inclui valores de índice repetidos. O argumento *ignore_index* do método `concat` elimina o problema.

```python
pd.concat([df1, df2], ignore_index=True)  # Gera uma cópia reindexada
```

Frequentemente, temos informações referentes às mesmas entidades (ex., pessoas, empresas, músicas, eventos) divididas em várias tabelas ou *DataFrames*. Embora o método `concat` possa ser usado para agregar colunas, ele não é a solução mais eficiente nesses casos.

```python
df2 = df1.copy().drop(["Ano", "Semestre"], axis = 1)
df2["CPF"] = ["000.000.00" + str(i) for i in range(7)]
df2["Idade"] = np.random.randint(18, 30, size = 7)
df2
```

Se usarmos o método `concat`, a coluna "Nome" é repetida.

```python
pd.concat([df1, df2], axis = 1)
```

Note que, se tentarmos usar `drop` para eliminar uma das colunas repetidas, as duas acabam sendo removidas.

```python
pd.concat([df1, df2], axis = 1).drop(["Nome"], axis = 1)
```

O método `merge` é uma melhor opção para agrupar colunas de diferentes *DataFrames* que compartilhar de uma ou mais colunas que identificam seus registros (linhas).

```python
df3 = pd.merge(left = df1, right = df2, on = "Nome")
df3
```

No exemplo anterior, usamos um tipo de junção (*join*) de *DataFrames* chamado *inner join*. O método `merge` inclui outras opções. Para mais informações, consulte a página [JOIN IN PANDAS: MERGE DATA FRAMES (INNER, OUTER, RIGHT, LEFT JOIN) IN PANDAS PYTHON"](https://www.datasciencemadesimple.com/join-merge-data-frames-pandas-python/).

# Formatos *long* e *wide*  

Alguns tipos de dados são organizados em *DataFrames* de forma que o campo ou variável que identifica das linhas apareça numa coluna, as variáveis de interesse sejam listadas numa única coluna, e os seus valores correspondentes sejam incluídos numa terceira coluna. Esse tipo de arranjo é conhecido por *formato longo* (*long*). 

A forma como organizamos nossos dataframes até agora é conhecida por *formado largo* (*wide*).

Podemos migrar do formato longo para o largo e vice-versa usando os métodos `pivot` e `melt`.

```python
# DataFrame original:
df3
```

```python
# Formato longo (long):
df3_longo = df3 \
    .melt(df3,
          id_vars = ["CPF"],
          var_name = "Variáveis",
          value_name = "Valores"
         ) \
    .sort_values("CPF")

df3_longo
```

```python
# Formato largo (wide):
df3_largo = df3_longo \
    .pivot(index = "CPF",
           columns = "Variáveis",
           values = "Valores") \
    .reset_index()
df3_largo
```

```python
df3_largo.reindex(columns = ["CPF", "Nome", "Idade", "Ano", "Semestre"],
                  copy = False)
```
